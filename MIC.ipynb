```python
#@title download from url

#########################################working dir
mainTempFolder="tmp/"
textOnlyFolder=mainTempFolder+"textOnly/"
inpaintedFolder=mainTempFolder+"inpainted/"
transalatedFolder=mainTempFolder+"translated/"

#remove if exist
!rm -r -f gallery-dl            #downloaded jpg
!rm -r -f $mainTempFolder       #worked jpg

#create working dir
for filePath in [textOnlyFolder,inpaintedFolder,transalatedFolder]:
  if not os.path.exists(filePath):
      os.makedirs(filePath)


#############################################download jpg from site
!gallery-dl $url
downloadFileList=glob.glob("gallery-dl/*/*/*.jpg")
downloadFileList.sort()
mangaName=os.path.basename(glob.glob("gallery-dl/*/*")[0])
print(mangaName)
print(downloadFileList)
print(os.path.basename(downloadFileList[0]))
PIL.Image.open(downloadFileList[0])
```

```python

#reisze image to make even num pixcel
#segmentation model conflict when odd num pixel
for i,imgPath in enumerate(tqdm(downloadFileList)):
    fileName=os.path.basename(imgPath)

    src = cv2.imread(imgPath)
    dst = cv2.resize(src, dsize=(src.shape[1]-(src.shape[1]%2),src.shape[0]-(src.shape[0]%2)), interpolation=cv2.INTER_LINEAR)
    cv2.imwrite(imgPath, dst)

#segmenation
maskDict=dict()
for i,imgPath in enumerate(tqdm(downloadFileList)):    
    fileName=os.path.basename(imgPath)

    #pred mask
    im = open_image(imgPath)
    pred = learner.predict(im)[0]
    mask = unpad_tensor(pred.px, im.px.shape)[0]
    mask = binary_dilation(mask, selem=disk(5))       #get bigger mask
    maskDict[fileName]=mask

```

```python
def inpaintSickZil2020(imgPath,OutPath,mask):      #https://github.com/KUR-creative/SickZil-Machine
    mask255= mask.astype(np.uint8) * 255
    maskImage=cv2.merge((mask255,mask255,mask255))
    oriImage = cv2.imread(imgPath)
    inpaintedImage = core.inpainted(oriImage, maskImage)        #notext image
    cv2.imwrite(OutPath, inpaintedImage)


def inpaintTelea2004(imgPath,OutPath,mask):      #https://docs.opencv.org/3.4/df/d3d/tutorial_py_inpainting.html  
    maskImage =mask.astype(np.uint8)*255
    oriImage = cv2.imread(imgPath)
    inpaintedImage = cv2.inpaint(oriImage,maskImage,3,cv2.INPAINT_TELEA)
    cv2.imwrite(OutPath, inpaintedImage)


def inpaintBiharmonic2018(imgPath,OutPath,mask):    #https://scikit-image.org/docs/dev/auto_examples/filters/plot_inpaint.html
    oriImage = io.imread(imgPath)
    inpaintedImage = inpaint.inpaint_biharmonic(oriImage, mask, multichannel=True)
    io.imsave(OutPath,inpaintedImage)


def inpaintPatchMatch2010(imgPath,OutPath,mask):  #https://github.com/vacancy/PyPatchMatch.git
    oriImage = cv2.imread(imgPath)
    inpaintedImage = patch_match.inpaint(oriImage, mask.astype(np.uint8), patch_size=5)
    cv2.imwrite(OutPath, inpaintedImage)




#inpaint mask
for i,imgPath in enumerate(tqdm(downloadFileList)):    
    fileName=os.path.basename(imgPath)
    mask=maskDict[fileName]

    #get inpaint
    if inpaintCode=="SickZil":
        inpaintSickZil2020(imgPath,inpaintedFolder+fileName,mask)
    elif inpaintCode=="Telea":
        inpaintTelea2004(imgPath,inpaintedFolder+fileName,mask)
    elif inpaintCode=="Biharmonic":
        inpaintBiharmonic2018(imgPath,inpaintedFolder+fileName,mask)
    elif inpaintCode=="PatchMatch":
        inpaintPatchMatch2010(imgPath,inpaintedFolder+fileName,mask)


    inpaintedImage = cv2.imread(inpaintedFolder+fileName)
    #get textOnlyImage
    oriImage = cv2.imread(imgPath)
    maskImage =np.array( mask.astype(np.uint8) )*255    
    textOnlyImage = cv2.bitwise_and(oriImage,oriImage,mask = maskImage)
    textOnlyImage[maskImage==0] = 255         
    cv2.imwrite(textOnlyFolder+fileName, textOnlyImage)

    #display
    if i==int(len(downloadFileList)/3):
      # cv2_imshow(oriImage)
      # cv2_imshow(maskImage)
      cv2_imshow(inpaintedImage)
      cv2_imshow(textOnlyImage)
```
